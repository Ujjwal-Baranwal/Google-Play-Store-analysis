Replacing nan rating values with their categorical mean values


```python
list1 = test.Category.sort_values().unique()
list2 = np.arange(0,33,1)
for i,j in zip(list1,list2):
    test.loc[test[(test['Rating'].isna())&(test['Category']==i)].Rating.index.tolist(),'Rating'] = test.groupby('Category').Rating.mean()[j]
 #   test.loc[test[(test['Rating'].isna())&(test['Category']=='ART_AND_DESIGN')].Rating.index.tolist(),'Rating'] = 0
```
